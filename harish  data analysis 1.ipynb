```python
import pandas as pd
import statsmodels.api as sm

# Convert categorical columns to numeric using one-hot encoding
df_encoded = pd.get_dummies(df[['Fuel_Type', 'Seller_Type', 'Transmission']], drop_first=True)

# Ensure numeric columns are of appropriate type
df[['Year', 'Kms_Driven', 'Owner']] = df[['Year', 'Kms_Driven', 'Owner']].apply(pd.to_numeric, errors='coerce')

# Combine all features into a single DataFrame
X = pd.concat([df[['Year', 'Kms_Driven', 'Owner']], df_encoded], axis=1)
X = sm.add_constant(X)  # Add an intercept for the regression model

# Dependent variable
y = df['Cost_Depreciation']

# Check for NaN values in X and y and handle them
X = X.dropna()
y = y[X.index]  # Ensure y matches the filtered X

# Explicitly convert all columns in X and y to numeric type
X = X.apply(pd.to_numeric, errors='coerce')
y = pd.to_numeric(y, errors='coerce')

# Check for any remaining non-numeric columns in X
non_numeric_cols = X.select_dtypes(exclude=['number']).columns
if non_numeric_cols.any():
    print("Non-numeric columns in X:", non_numeric_cols)
    # Handle these non-numeric columns appropriately (e.g., convert to numeric, drop them, etc.)
    # Example: Dropping non-numeric columns
    X = X.drop(non_numeric_cols, axis=1)

# Fit the model
model = sm.OLS(y, X).fit()

# Print the summary
print(model.summary())
```
